```python
import pandas as pd
```

```python
def load_data_to_dataframe(file_path):
    return pd.read_csv(file_path, sep=';')
```

```python
monster_data = load_data_to_dataframe('monstermanual.txt')
monster_data.drop_duplicates()
monster_data = monster_data.set_index('Name')
# Convert the 'Level/CR' column to numeric, handling fractions like '1/2', '1/4', etc.
monster_data['ChallengeRating'] = pd.to_numeric(monster_data['ChallengeRating'], errors='coerce')
# Sort the DataFrame by 'Level/CR' in descending order
monster_data_hardest = monster_data.sort_values(by='ChallengeRating', ascending=False)
monster_data_hardest = monster_data_hardest.head(10)
monster_data_hardest[['HP', 'AC', 'ChallengeRating', 'ChallengeXP']]
```

```python
monster_data_ac = monster_data.sort_values(by='AC', ascending=False)
monster_data_ac = monster_data_ac.head(10)
monster_data_ac[['HP', 'AC', 'ChallengeRating', 'ChallengeXP']]
```

```python
monster_data_hp = monster_data.sort_values(by='HP', ascending=False)
monster_data_hp = monster_data_hp.head(10)
monster_data_hp[['HP', 'AC', 'ChallengeRating', 'ChallengeXP']]
```
